```r
a=rnorm(100,0,1)
b=rnorm(100,2,1)
```

```r
cor(a,b)
```

```r
#png('~/Documents/blog/correlation.example.png',height=700, width=2000, pointsize = 40)
par(mfrow=c(1,3))
plot(b, b/runif(100,1,7), xlab="Water Consumption (Cups)", ylab="Productivity (Hours Worked)", col="palevioletred2", 
     main=paste("cor = ",round(cor(b, b/runif(100,1,100)),2), sep=""))

plot(b, b/runif(100,1,2), xlab="Coffee Consumption (Cups)", ylab="Productivity (Hours Worked)", col="palevioletred2", 
     main=paste("cor = ",round(cor(b, b/runif(100,1,3)),2), sep=""))

plot(b, -b/runif(100,1,2), xlab="Beer Consumption (Bottles)", ylab="Productivity (Hours Worked)", col="dodgerblue2", 
     main=paste("cor = ",round(cor(b, -b/runif(100,1,2)),2), sep=""), yaxt='n')
axis(side=2, at=-4:0, labels=0:4)
#dev.off()
```
